## Postprocessing

```python
import numpy as np

extracted_cols = [col for col in df.columns if col.startswith('extracted_number_')]

df['extract_numbers'] = df[extracted_cols].apply(
    lambda row: ','.join(str(x) for x in row if pd.notna(x)),
    axis=1
)

# Group by file and combine multiple rows per file, deduplicating numbers
df_grouped = df.groupby('file')['extract_numbers'].agg(
    lambda s: ','.join(
        dict.fromkeys(
            [t.strip() for t in ','.join(s).split(',') if t.strip()]
        )
    )
).reset_index()


df_grouped['extract_numbers'] = df_grouped['extract_numbers'].replace('', np.nan)
df_grouped.head(10)

```

```python
df_grouped[df_grouped.extract_numbers.str.contains(',', na=False)]
```

```python
import json
with open('../data/labels.json', 'r') as f:
    labels = json.load(f)

# Convert dictionary to DataFrame
label_df = pd.DataFrame(list(labels.items()), columns=['file', 'label'])

result_df = pd.merge(df_grouped, label_df, on='file', how='left')
result_df.head(10)
```

```python
# Check where labels are null
result_df[result_df['label'].isnull()]
```

```python
ct_correct = (result_df['extract_numbers'] == result_df['label']).sum()
print(f"Correct: {ct_correct} / {len(result_df)}")
print(f"Accuracy: {ct_correct / len(result_df):.2%}")

# count how many 'ambiguous' labels there are
ambiguous_count = sum(1 for label in labels.values() if 'ambiguous' in label)
print(f"Number of ambiguous labels: {ambiguous_count} / {len(result_df)}")

# check where it was wrong and extract_numbers is not null
print("Wrong predictions where both label and extract_numbers are non-null:")
wrong_df = result_df[(result_df['extract_numbers'] != result_df['label']) & (result_df['extract_numbers'].notnull()) & (result_df['label'].notnull())]
wrong_df
```

## Validating Heuristics

1. Labels must be > 3 characters 
2. Labels cannot contain no numbers
3. Any extracted label that have multiple values

```python
# filter result_df for extract_numbers that are <= 3 characters and non null
filtered_df = result_df[result_df['extract_numbers'].notnull() & (result_df['extract_numbers'].str.len() <= 3)]
filtered_df # X51005719899 is an anomaly
```

```python
# filter result df for extract_numbers that contain no numbers
filtered_df = result_df[result_df['extract_numbers'].notnull() & (~result_df['extract_numbers'].str.contains(r'\d', na=False))]
filtered_df
```

```python
# filter result_df for extract_numbers that contain a comma
filtered_df = result_df[result_df['extract_numbers'].notnull() & (result_df['extract_numbers'].str.contains(',', na=False))]
filtered_df
```
