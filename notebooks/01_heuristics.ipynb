```python
import pandas as pd

df = pd.read_csv('../data/invoices.csv')
df.head()
```

## Testing Heuristics

```python
df['extracted_number_INV#'] = df['text'].str.extract(r'INV#\s*([^\s]+)')
df[df['extracted_number_INV#'].notnull()]

# groupby file and check if there are multiple different extracted numbers
df.groupby('file')['extracted_number_INV#'].nunique().reset_index().sort_values(by='extracted_number_INV#', ascending=False).head(20)

# check file X51006912972.jpg
# df[df['file'] == 'X51006912972.jpg']
```

```python
# Heuristic 1: INV#
df['extracted_number_INV#'] = df['text'].str.extract(r'INV#\s*([^\s]+)')
df[df['extracted_number_INV#'].notnull()]
print(f'Heuristic "INV#": {len(df[df['extracted_number_INV#'].notnull()])} / {len(df)}')

# Heuristic 2: INV:
df['extracted_number_INV'] = df['text'].str.extract(r'INV:\s*([^\s]+)')
df[df['extracted_number_INV'].notnull()]
print(f'Heuristic "INV:": {len(df[df['extracted_number_INV'].notnull()])} / {len(df)}')

# Heuristic 3: INV-NO.
df['extracted_number_INV-NO'] = df['text'].str.extract(r'INV-NO\.\s*([^\s]+)')
df[df['extracted_number_INV-NO'].notnull()]
print(f'Heuristic "INV-NO": {len(df[df['extracted_number_INV-NO'].notnull()])} / {len(df)}')

# Heuristic 4: INV NO
df['extracted_number_INV_NO'] = df['text'].str.extract(r'INV\s*NO\.?\s*[:\-]?\s*([A-Za-z0-9/\-]+)')
df[df['extracted_number_INV_NO'].notnull()]
print(f'Heuristic "INV NO": {len(df[df['extracted_number_INV_NO'].notnull()])} / {len(df)}')



# Heuristic 5: Slip
df['extracted_number_Slip'] = df['text'].str.extract(r'SLIP\s*([^\s]+)')
df[df['extracted_number_Slip'].notnull()]
print(f'Heuristic "SLIP": {len(df[df['extracted_number_Slip'].notnull()])} / {len(df)}')

# Heurstic 6: INVOICE NO
df['extracted_number_INVOICE_NO'] = df['text'].str.extract(r'INVOICE\s*NO\s*[:\-]?\s*([A-Za-z0-9/\-]+)')
df[df['extracted_number_INVOICE_NO'].notnull()]
print(f'Heuristic "INVOICE NO": {len(df[df['extracted_number_INVOICE_NO'].notnull()])} / {len(df)}')

# Heuristic 7: RECEIPT NO.
df['extracted_number_RECEIPT_NO'] = df['text'].str.extract(r'RECEIPT\s*NO\s*[:\-]?\s*([A-Za-z0-9/\-]+)')
df[df['extracted_number_RECEIPT_NO'].notnull()]
print(f'Heuristic "RECEIPT NO": {len(df[df['extracted_number_RECEIPT_NO'].notnull()])} / {len(df)}')

# Heuristic 8: CB# : 
df['extracted_number_CB#'] = df['text'].str.extract(r'CB# :\s*([A-Za-z0-9/\-]+)')
df[df['extracted_number_CB#'].notnull()]
print(f'Heuristic "CB# :": {len(df[df['extracted_number_CB#'].notnull()])} / {len(df)}')

# Heursitic 9: BILL NO
df['extracted_number_BILL_NO'] = df['text'].str.extract(r'BILL\s*NO\s*[:\-]?\s*([A-Za-z0-9/\-]+)')
df[df['extracted_number_BILL_NO'].notnull()]
print(f'Heuristic "BILL NO": {len(df[df['extracted_number_BILL_NO'].notnull()])} / {len(df)}')

# Heuristic 10: INVOICE #
df['extracted_number_INVOICE#'] = df['text'].str.extract(r'INVOICE\s*#\s*:?[\s]*([A-Za-z0-9/\-]+)')
df[df['extracted_number_INVOICE#'].notnull()]
print(f'Heuristic "INVOICE#": {len(df[df['extracted_number_INVOICE#'].notnull()])} / {len(df)}')   

# Heuristic 11: C/N NO:
df['extracted_number_CN_NO'] = df['text'].str.extract(r'C/N\s*NO\s*[:\-]?\s*([A-Za-z0-9/\-]+)')
df[df['extracted_number_CN_NO'].notnull()]
print(f'Heuristic "C/N NO": {len(df[df['extracted_number_CN_NO'].notnull()])} / {len(df)}')

# Heuristic 12: TRANSACTION NO:
df['extracted_number_TRANSACTION_NO'] = df['text'].str.extract(r'TRANSACTION\s*NO\s*[:\-]?\s*([A-Za-z0-9/\-]+)')
df[df['extracted_number_TRANSACTION_NO'].notnull()]
# print(f'Heuristic "TRANSACTION NO": {len(df[df['extracted_number_TRANSACTION_NO'].notnull()])} / {len(df)}')



```

```python
extracted_cols = [col for col in df.columns if col.startswith('extracted_number_')]

df['extract_numbers'] = df[extracted_cols].apply(
    lambda row: [x for x in row if pd.notna(x)],
    axis=1
)

# Group by file and flatten lists
df_grouped = df.groupby('file')['extract_numbers'].agg(
    lambda lists: list(dict.fromkeys([item for sublist in lists for item in sublist]))
).reset_index()
```

```python
# give me rows where df_grouped['extract_numbers'] has more than 1 value
df_grouped[df_grouped['extract_numbers'].apply(lambda x: len(x) > 1)]
```
